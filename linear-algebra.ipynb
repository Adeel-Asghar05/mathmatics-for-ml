```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Generate sample dataset
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
print(X)

# Plot original data
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
plt.title("Original Data")
plt.show()

# Perform PCA
pca = PCA(n_components=2)
pca.fit(X)

# Function to draw PCA vectors
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops = dict(arrowstyle='->', linewidth=2, shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# Plot data with principal components
plt.scatter(X[:, 0], X[:, 1])
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)

plt.axis('equal')
plt.title("PCA Components")
plt.show()

# PCA with 1 component
pca_1 = PCA(n_components=1)
pca_1.fit(X)
X_pca = pca_1.transform(X)
X_new = pca_1.inverse_transform(X_pca)
print ("newx" ,X_new)

# Plot original vs reduced reconstruction
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8, label='Reconstructed')
plt.axis('equal')
plt.title("Original vs Reconstructed (1D PCA)")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA

# Sample dataset (4x3)
X = np.array([
    [2, 3, 1],
    [4, 7, 2],
    [5, 8, 3],
    [3, 5, 2]
])

print("Original Data:\n", X)

# Run PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("\nTransformed Data (2 PCs):\n", X_pca)
print("\nExplained Variance Ratio:", pca.explained_variance_ratio_)
print("\nComponents (Eigenvectors):\n", pca.components_)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#  1. Load Dataset
df = pd.read_csv('BTC-USD_with_Indicators.csv')

#  2. Remove time/date column(s)
# Adjust column name if needed
time_columns = ['OpenTime']  # modify if name is different
df = df.drop(columns=[col for col in time_columns if col in df.columns])

#  3. Keep only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

print("Original numeric shape:", numeric_df.shape)

#  4. Scale data (required for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

#  5. Apply PCA (keep 95% variance)
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(scaled_data)

print("Reduced shape:", X_pca.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance kept:", sum(pca.explained_variance_ratio_))

#  6. Create DataFrame for PCA output
df_pca = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(X_pca.shape[1])])

#  7. Save to CSV
output_path = '/mnt/data/pca_indicator.csv'
df_pca.to_csv(output_path, index=False)

print(f"PCA data saved to: {output_path}")
print(df_pca.head())
```
